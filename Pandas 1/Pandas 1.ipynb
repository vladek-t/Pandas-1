### Задание 10

Определить минимальное, максимальное, среднее и медианное значение индекса экстраверсии в датафрейме `pure`. Сохранить полученные результаты в отдельные переменные (их должно быть 4).

Добавить в датафрейм `pure` столбец `high`, состоящий из 0 и 1, где 1 соответствует респондентам, уровень экстраверсии которых выше значения $m = \max\{\text{median}, \text{mean}\}$, то есть максимума из медианного и среднего значения, а 0 — респондентам с уровнем экстраверсии не выше $m$.

```python
import pandas as pd
import numpy as np

ps = pd.read_csv('extraversion.csv', encoding='UTF-8')

columns = ps.columns.to_list()

columns[4:] = [column.replace(' ', '') for column in columns[4:]]
ps.columns = columns

extra_yes = ps.loc[:, ['Q1', 'Q3', 'Q8', 'Q10', 'Q13', 'Q17', 'Q22', 'Q25', 'Q27', 'Q39', 'Q44', 'Q46', 'Q49', 'Q53',
                       'Q56']]

extra_no = ps.loc[:, ['Q5', 'Q15', 'Q20', 'Q29', 'Q32', 'Q34', 'Q37', 'Q41', 'Q51']]

extra_yes_sum = extra_yes.isin(['Да']).apply(np.sum, axis=1)

extra_no_sum = extra_no.isin(['Нет']).apply(np.sum, axis=1)

ps['extra'] = extra_no_sum + extra_yes_sum

s = pd.Series(ps.loc[:, 'sex'])

cond = pd.Series(s == 'Женский')
cond = cond.astype(int)

ps['female'] = cond

pure = ps[(ps['volunteer'] == 'Да') & (ps['extra'] > 15) | (ps['volunteer'] == 'Нет') & (ps['extra'] < 15)]

volunteer = 0
not_volunteer = 0

for x in pure.loc[:, 'volunteer']:
    if x == 'Да':
        volunteer += 1
    else:
        not_volunteer += 1

min_extra = pure.loc[:, 'extra'].agg('min')
max_extra = pure.loc[:, 'extra'].agg('max')
mean_extra = pure.loc[:, 'extra'].agg('mean')
median_extra = pure.loc[:, 'extra'].agg('median')

d = []
for x in pure['extra']:
    if x > median_extra and x > mean_extra:
        x = 1
        d.append(x)
    else:
        x = 0
        d.append(x)

pure['high'] = d

pure
```
